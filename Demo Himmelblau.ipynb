# Demo: Himmelblau

In this demonstration we will be fitting Himmelblau's function. This is defined by:

$f(x,y) = (x^2 + y - 11)^2 + (x + y^2 - 7)^2$

There are four local minima where f(x, y) = 0.

These are at:
- (3, 2)
- (-2.805118, 3.131312)
- (-3.779310, -3.283186)
- (3.584428, -1.848126)

```python
import gaga as ga
import numpy as np
from matplotlib import rc
rc('animation', html='html5')
```

# Define the genes

```python
genes = {'x':(-5, 5),
        'y':(-5, 5)}
```

# Define fitness function

Our goal is to locate the minimas, so the fitness score is $f(x, y)$

```python
def evaluate(individual):

    #  unpack chromosome
    x = individual.CHROMOSOME['x']
    y = individual.CHROMOSOME['y']

    individual.FITNESS_SCORE = pow(pow(x, 2) + y - 11, 2) + pow(x + pow(y, 2) - 7, 2)
```

# Set up and run the genetic algorithm

```python
np.random.seed(1)
sim = ga.ga(genes, 
            evaluate,
            population_size = 100,
            epoch = 50,
            mutate_crossover = True,
            speciate = True,
            mutate = 0.4,
            sigma = 0.5,
            results_folder = "demo-himmelblau")
sim.run_simulation()
```

# Results

```python
sim.results.plot_fitness()
```
