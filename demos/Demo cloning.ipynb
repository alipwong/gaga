# The power of cloning

To demonstrate the power of cloning we will use Rosenbrock's banana function.

This is defined by:

$f(x,y) = (a-x)^2 + b(y - x^2)^2$

For the constants $a = 1$ and $b = 100$, the minimum is at (1,1).

$f(1,1) = 0$

Image source:[wikipedia](https://en.wikipedia.org/wiki/Rosenbrock_function) 

```python
import gaga as ga
from matplotlib import rc
rc('animation', html='html5')

results_folder = 'demo-cloning'

genes = {'x':(-2, 2),
        'y':(-1, 3)}

def evaluate(individual):

    # constants
    a = 1
    b = 100

    #  unpack chromosome
    x = individual.genes['x']
    y = individual.genes['y']

    individual.fitness_score = (pow(a - x, 2) + b * pow(y - pow(x, 2), 2))
```

# Without cloning

```python
seed = 1
sim = ga.ga(genes, 
            evaluate,
            clone = 0,
            mutate = 0.6,
            sigma = 0.3,
            population_size = 10,
            results_folder = results_folder)
```

```python
sim.run_simulation(seed = seed)
anim = sim.results.animate('x', 'y', 
                           bounds = [-2, 2, -1, 3], s = 20, 
                           alpha = 0.25, optimum = [1, 1], fmax = 100, 
                           filename = "no_cloning_{}".format(seed))
```

```python
sim.results.plot_fitness(fname = "no_cloning_{}".format(seed))
```

# With cloning

```python
seed = 1
sim = ga.ga(genes, 
            evaluate,
            clone = 1,
            mutate = 0.6,
            sigma = 0.3,
            population_size = 10,
            results_folder = results_folder)
```

```python
sim.run_simulation(seed = seed)
anim = sim.results.animate('x', 'y', 
                           bounds = [-2, 2, -1, 3], s = 20, 
                           alpha = 0.25, optimum = [1, 1], fmax = 100, 
                           filename = "cloning_{}".format(seed))
```

```python
sim.results.plot_fitness(fname = "cloning_{}".format(seed))
```
