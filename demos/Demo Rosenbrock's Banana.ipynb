# Demo: Rosenbrock's Banana

In this demonstration we will be fitting Rosenbrock's function. This is defined by:

$f(x,y) = (a-x)^2 + b(y - x^2)^2$

For the constants $a = 1$ and $b = 100$, the minimum is at (1,1).

$f(1,1) = 0$

```python
import gaga as ga
import numpy as np
from matplotlib import rc
rc('animation', html='html5')
```

# Define the genes

```python
genes = {'x':(-2, 2),
        'y':(-1, 3)}
```

# Define fitness function


Our goal is to locate the minimum, so the fitness score is $f(x, y)$

```python
def evaluate(individual):

    # constants
    a = 1
    b = 100

    #  unpack chromosome
    x = individual.genes['x']
    y = individual.genes['y']

    individual.fitness_score = (pow(a - x, 2) + b * pow(y - pow(x, 2), 2))
```

# Set up the genetic algorithm

```python
np.random.seed(1)
sim = ga.ga(genes, 
            evaluate, 
            population_size = 50,
            results_folder = "demo-rosenbrock")
sim.run_simulation()
```

# Results

```python
sim.results.plot_fitness()
```

```python
anim = sim.results.animate('x', 'y', optimum = [1, 1], inset = [0.75, 1.25, 0.75, 1.25], fmax = 100)
```
