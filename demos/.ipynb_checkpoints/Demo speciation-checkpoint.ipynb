# Demo: speciation


For further reference see: [here](https://awon8465.github.io/gagadoc/tuning-the-ga/speciation.html)

In this demonstration we will be fitting Himmelblau's function. This is defined by:

$f(x,y) = (x^2 + y - 11)^2 + (x + y^2 - 7)^2$

There are four local minima where f(x, y) = 0.

These are at:
- (3, 2)
- (-2.805118, 3.131312)
- (-3.779310, -3.283186)
- (3.584428, -1.848126)

### Speciation

```python
import gaga as ga
import numpy as np
from matplotlib import rc
rc('animation', html='html5')

genes = {'x':(-5, 5),
        'y':(-5, 5)}

def evaluate(individual):

    #  unpack chromosome
    x = individual.genes['x']
    y = individual.genes['y']

    individual.fitness_score = pow(pow(x, 2) + y - 11, 2) + pow(x + pow(y, 2) - 7, 2)
```

```python
seed = 1
sim = ga.ga(genes, 
                evaluate,
                population_size = 400,
                epoch = 50,
                mutate_crossover = True,
                speciate = True,
                mutate = 0.6,
                winrate = 1,
                sigma = 0.5,
                results_folder = "demo-speciation")
sim.run_simulation(seed = seed)
anim = sim.results.animate('x', 'y', alpha = 0.1, fmax = 100, repeat = False, filename = "speciate_{}".format(seed))
```
