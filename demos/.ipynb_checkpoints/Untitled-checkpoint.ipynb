```python
import gaga as ga
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D 
rc('animation', html='html5')
```

## Easom

```python
# Visualising the Easom function
delta = 0.01
x = np.arange(-5, 7, delta)
y = np.arange(-5, 7, delta)
X, Y = np.meshgrid(x, y)

exponent = np.power(X - np.pi, 2) + np.power(Y - np.pi, 2)
Z = -np.cos(X)* np.cos(Y) * np.exp(-exponent)

fig = plt.figure(figsize = (10,10))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, -Z)
```

```python
def evaluate(individual):

    #  unpack genes
    x = individual.genes['x']
    y = individual.genes['y']
    
    exponent = np.power(x - np.pi, 2) + np.power(y - np.pi, 2)
    individual.fitness_score = -np.cos(x)* np.cos(y) * np.exp(-exponent)
    
genes = {'x':(-20, 20),
        'y':(-20, 20)}
```

## Simple Easom, no cloning

```python
sim = ga.ga(genes, 
            evaluate,
            population_size = 50,
            clone = 0,
            winrate = 0.65,
            results_folder = "MLDI-assignment")
sim.run_simulation(seed = 0)
sim.results.print_best()
anim = sim.results.animate('x', 'y', optimum = [np.pi, np.pi], inset = True, filename = "easom-clone")
sim.results.plot_fitness()
```

```python
sim = ga.ga(genes, 
            evaluate,
            population_size = 50,
            clone = 0,
            winrate = 0.65,
            results_folder = "MLDI-assignment")
sim.run_simulation(seed = 1)
sim.results.print_best()
anim = sim.results.animate('x', 'y', optimum = [np.pi, np.pi], inset = True, filename = "easom-clone")
sim.results.plot_fitness()
```

## Double Easom

```python
# Visualising the double Easom function
delta = 0.01
x = np.arange(-5, 17, delta)
y = np.arange(-5, 17, delta)
X, Y = np.meshgrid(x, y)

exponent = np.power(X - np.pi, 2) + np.power(Y - np.pi, 2)
Z1 = -np.cos(X)* np.cos(Y) * np.exp(-exponent)

exponent = np.power(X, 2) + np.power(Y, 2)
Z2 = -0.5 * np.cos(X+np.pi)* np.cos(Y+np.pi) * np.exp(-exponent)

fig = plt.figure(figsize = (10,10))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, -(Z1 + Z2))
```

```python
def evaluate(individual):

    #  unpack genes
    x = individual.genes['x']
    y = individual.genes['y']
    
    exponent = np.power(x - np.pi, 2) + np.power(y - np.pi, 2)
    Z1 = -np.cos(x)* np.cos(y) * np.exp(-exponent)

    exponent = np.power(x, 2) + np.power(y, 2)
    Z2 = -0.5 * np.cos(x+np.pi)* np.cos(y+np.pi) * np.exp(-exponent)
    
    individual.fitness_score = Z1 + Z2
    
genes = {'x':(-5, 5),
        'y':(-5, 5)}
```

### No cloning

```python
sim = ga.ga(genes, 
            evaluate, 
            population_size = 20,
            clone = 0,
            results_folder = "MLDI-assignment")

sim.run_simulation(seed = 0)
anim = sim.results.animate('x', 'y', optimum = [[0, np.pi], [0, np.pi]], inset = [-1, 4, -1, 4], filename = "double-easom-no-clone")
sim.results.plot_fitness()
```

```python

sim = ga.ga(genes, 
            evaluate, 
            population_size = 20,
            clone = 0,
            results_folder = "MLDI-assignment")

sim.run_simulation(seed = 1)
anim = sim.results.animate('x', 'y', optimum = [[0, np.pi], [0, np.pi]], inset = [-1, 4, -1, 4], filename = "double-easom-no-clone")
sim.results.plot_fitness()
```

## With noise

```python
def evaluate(individual):

    #  unpack genes
    x = individual.genes['x']
    y = individual.genes['y']
    
    exponent = np.power(x - np.pi, 2) + np.power(y - np.pi, 2)
    Z1 = -np.cos(x)* np.cos(y) * np.exp(-exponent)

    exponent = np.power(x, 2) + np.power(y, 2)
    Z2 = -0.5 * np.cos(x+np.pi)* np.cos(y+np.pi) * np.exp(-exponent)
    
    individual.fitness_score = Z1 + Z2 + np.random.normal(scale = 0.4)

genes = {'x':(-5, 5),
        'y':(-5, 5)}
    
np.random.seed(1)
sim = ga.ga(genes, 
            evaluate, 
            population_size = 100,
            clone = 0,
            winrate = 0.5,
            sigma = 1,
            results_folder = "MLDI-assignment")
sim.run_simulation()

anim = sim.results.animate('x', 'y', optimum = [[0, np.pi], [0, np.pi]], inset = [-1, 4, -1, 4], filename = "double-easom-no-clone")
```

```python
sim = ga.ga(genes, 
            evaluate, 
            population_size = 20,
            clone = 2,
            winrate = 0.7,
            results_folder = "MLDI-assignment")
sim.run_simulation(seed = 0)

anim = sim.results.animate('x', 'y', optimum = [[0, np.pi], [0, np.pi]], inset = [-1, 4, -1, 4], filename = "double-easom-no-clone")
```

## Equal minima

```python
def evaluate(individual):

    #  unpack genes
    x = individual.genes['x']
    y = individual.genes['y']
    
    exponent = np.power(x - np.pi, 2) + np.power(y - np.pi, 2)
    Z1 = -np.cos(x)* np.cos(y) * np.exp(-exponent)

    exponent = np.power(x, 2) + np.power(y, 2)
    Z2 = -np.cos(x+np.pi)* np.cos(y+np.pi) * np.exp(-exponent)
    
    individual.fitness_score = Z1 + Z2

genes = {'x':(-4, 7),
        'y':(-4, 7)}
 
```

```python
sim = ga.ga(genes, 
            evaluate, 
            population_size = 100,
            sigma = 0.5, 
            speciate = False,
            mutate = 0.6,
            winrate = 1,
            results_folder = "MLDI-assignment")
sim.run_simulation(seed = 0)

anim = sim.results.animate('x', 'y', optimum = [[0, np.pi], [0, np.pi]], inset = [-1, 4, -1, 4], filename = "double-easom-no-clone")
```
