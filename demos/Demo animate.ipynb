# Demo animate

This is a quick tutorial for using ``animate`` to

```python
import gaga as ga
import numpy as np
from matplotlib import rc
rc('animation', html='html5')

genes = {'x':(-5, 5),
        'y':(-5, 5)}

def evaluate(individual):

    x = individual.genes['x']
    y = individual.genes['y']

    individual.fitness_score = pow(pow(x, 2) + y - 11, 2) + pow(x + pow(y, 2) - 7, 2)

seed = 1
sim = ga.ga(genes, 
            evaluate,
            population_size = 200,
            epoch = 50,
            mutate_crossover = True,
            speciate = True,
            mutate = 0.4,
            sigma = 0.5,
            results_folder = "demo-animate")
sim.run_simulation(seed = seed)
```

```python
anim = sim.results.animate('x', 'y', filename = 'default')
```

## fmax and fmin

These can be used to specify the maximum and minimum values of the fitness score colorbar

```python
anim = sim.results.animate('x', 'y', 
                           fmax = 100,
                           filename = 'fmax')
```

## Mark the optimum

#### A single optimum

```python
anim = sim.results.animate('x', 'y', 
                           optimum = [3,2],
                           oc = 'w',
                           fmax = 100,
                           filename = 'single-optimum')
```

#### Multiple optimum

```python
anim = sim.results.animate('x', 'y', 
                           optimum = [[3, -2.805118, -3.779310, 3.584428], [2, 3.131312, -3.283186, -1.848126]],
                           oc = 'w',
                           fmax = 100,
                           filename = 'multiple-optimum')
```

## Inset

By default, the inset will be around the median individual of the final population. This should be where the population converges to.

```python
anim = sim.results.animate('x', 'y', 
                           optimum = [[3, -2.805118, -3.779310, 3.584428], [2, 3.131312, -3.283186, -1.848126]],
                           oc = 'w',
                           fmax = 100,
                           inset = True,
                           filename = 'default_inset')
```

#### You can also specify the location of the inset

```python
anim = sim.results.animate('x', 'y', 
                           optimum = [[3, -2.805118, -3.779310, 3.584428], [2, 3.131312, -3.283186, -1.848126]],
                           fmax = 100,
                           oc = 'w',
                           inset = [-3.5, -2, 2.3, 3.8],
                           filename = 'specified_inset')
```

## Fitness on a logscale

```python
anim = sim.results.animate('x', 'y', 
                           optimum = [[3, -2.805118, -3.779310, 3.584428], [2, 3.131312, -3.283186, -1.848126]],
                           oc = 'w',
                           inset = True,
                           log_scale = True,
                           fmin = 1,
                           filename = 'logscale')
```

## Transparency

Increasing the transparency can help you see which areas your population spends the most time exploring.

```python
nim = sim.results.animate('x', 'y', 
                           optimum = [[3, -2.805118, -3.779310, 3.584428], [2, 3.131312, -3.283186, -1.848126]],
                           oc = 'w',
                           fmax = 100,
                           alpha = 0.05,
                          s = 10,
                           filename = 'alpha')
```

## Change the colormap

```python
from matplotlib import cm


anim = sim.results.animate('x', 'y', 
                           cmap = cm.viridis_r,
                           fmax = 100,
                           filename = 'cmap')
```
