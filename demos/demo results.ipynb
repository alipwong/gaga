# Demo results

This is just a quick walkthrough on how to access the results of your simulation

```python
import gaga as ga
from matplotlib import rc
rc('animation', html='html5')

genes = {'x':(-2, 2),
        'y':(-1, 3)}

def evaluate(individual):

    # constants
    a = 1
    b = 100

    #  unpack chromosome
    x = individual.genes['x']
    y = individual.genes['y']

    individual.fitness_score = (pow(a - x, 2) + b * pow(y - pow(x, 2), 2))
    

seed = 1
sim = ga.ga(genes, evaluate)

sim.run_simulation(seed = seed)
```

## Accessing the results directly from the simulation

```python
anim = sim.results.animate('x', 'y')
```

```python
sim.results.plot_fitness()
```

## Accessing the results from a saved object

After running the simulation, a results folder will be created with the following structure:

``
results
|    history.txt
|    results_obj
``

We can load ``results_obj`` directly

```python
import bz2
import pickle

with bz2.open("./results/results_obj", "rb") as f:
    results = pickle.load(f)
```

```python
anim = results.animate('x', 'y')
```
